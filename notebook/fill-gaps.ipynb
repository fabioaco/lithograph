```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def fill_intervales(data0):
    data = np.copy(data0)
    prev = np.nan
    for i in range(data.shape[0]):
        if ~np.isnan(data[i]):
            prev = data[i]
        else:
            data[i] = prev

    return data
```

```python
fname = os.path.abspath("../data/well_logs_v2.csv")
```

```python
df = pd.read_csv(fname)
df.tail()
```

```python
facies = fill_intervales(df.FACIES.values)
facies[facies == -999] = np.nan
df["FACIES"] = facies
df.tail()
```

```python
df.to_csv("../data/well_logs_v2_filled.csv")
```
